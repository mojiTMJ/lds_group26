### Assignment 1
Write a python program that splits fact.csv into three separate fact tables, one for each line of products: cpu, gpu and ram. Write your program using only basic python functionalities (the use of the pandas library is forbidden).

```python
import threading
import pyodbc
import csv

Fact_Gpu=[]
Fact_Cpu=[]
Fact_Ram=[]

server = 'tcp:apa.di.unipi.it'
username = 'group26' 
password = 'l3wwh' 
connectionString = 'DRIVER={ODBC Driver 17 for SQL Server};SERVER='+server+';UID='+username+';PWD='+ password


_db = pyodbc.connect(connectionString,autocommit=True)
_cursor = _db.cursor()

db = pyodbc.connect(connectionString,autocommit=True)
cursor = db.cursor()

__db = pyodbc.connect(connectionString,autocommit=True)
__cursor = __db.cursor()


cursor.execute("DELETE FROM [Group26HWMart].[group26].[ram_sales]")
_cursor.execute("DELETE FROM [Group26HWMart].[group26].[gpu_sales]")
__cursor.execute("DELETE FROM [Group26HWMart].[group26].[cpu_sales]")


def SetFactTables():
        print("adding fact datas to fact tables")
        print('ram size: ', len(Fact_Ram))
        print('gpu size: ', len(Fact_Gpu))
        print('cpu size: ', len(Fact_Cpu))
        def _setRam():
            j = 1
            cursor.fast_executemany = True
            tmp = 0
            size = int(len(Fact_Ram)/200000)
            for l in range(size):
                temp = (l + 1)*200000
                cursor.executemany("INSERT INTO [Group26HWMart].[group26].[ram_sales] VALUES (?, ?, ?, ?, ?, ?)", Fact_Ram[tmp:temp])
                tmp = temp
                print('thread ram ', j)
                j = j + 1
            cursor.executemany("INSERT INTO [Group26HWMart].[group26].[ram_sales] VALUES (?, ?, ?, ?, ?, ?)", Fact_Ram[tmp:])
            cursor.close()
            print("Ram Fact table splited!")
        
        def _setCpu():
            j = 1
            __cursor.fast_executemany = True
            tmp = 0
            size = int(len(Fact_Cpu)/200000)
            for l in range(size):
                temp = (l + 1)*200000
                __cursor.executemany("INSERT INTO [Group26HWMart].[group26].[cpu_sales] VALUES (?, ?, ?, ?, ?, ?)", Fact_Cpu[tmp:temp])
                tmp = temp
                print('thread cpu ', j)
                j = j + 1
            __cursor.executemany("INSERT INTO [Group26HWMart].[group26].[cpu_sales] VALUES (?, ?, ?, ?, ?, ?)", Fact_Cpu[tmp:])
            __cursor.close()
            print("Cpu Fact table splited!")

        def _setGpu():
            j = 1
            _cursor.fast_executemany = True
            tmp = 0
            size = int(len(Fact_Gpu)/200000)
            for l in range(size):
                temp = (l + 1) * 200000
                _cursor.executemany("INSERT INTO [Group26HWMart].[group26].[gpu_sales] VALUES (?, ?, ?, ?, ?, ?)", Fact_Gpu[tmp:temp])
                tmp = temp
                print('thread gpu ', j)
                j = j + 1
            _cursor.executemany("INSERT INTO [Group26HWMart].[group26].[gpu_sales] VALUES (?, ?, ?, ?, ?, ?)", Fact_Gpu[tmp:])
            _cursor.close()
            print("Gpu Fact table splited!")

        a=threading.Thread(target=_setRam)
        b=threading.Thread(target=_setGpu)
        c=threading.Thread(target=_setCpu)

        a.start()
        b.start()
        c.start()
        
        a.join()
        b.join()
        c.join()

#TASK NUMBER 1 : SPLIT FACT TABLES
def SplitFactTables():
    _reader = open("fact.csv")
    print("Splitting Fact Tables...")
    for row in csv.DictReader(_reader):
        if row["gpu_code"] not in (None, ''):
            Fact_Gpu.append((int(float(row["gpu_code"])),int(row["time_code"]), int(row["geo_code"]), int(row["vendor_code"]), float(row["sales_uds"]), row["sales_currency"] ))
            
        elif row["ram_code"] not in (None, ''):
            Fact_Ram.append((int(float(row["ram_code"])),int(row["time_code"]), int(row["geo_code"]), int(row["vendor_code"]), float(row["sales_uds"]), row["sales_currency"] ))

        elif row["cpu_code"] not in (None, ''):
            Fact_Cpu.append((int(float(row["cpu_code"])),int(row["time_code"]), int(row["geo_code"]), int(row["vendor_code"]), float(row["sales_uds"]), row["sales_currency"] ))
    SetFactTables()

    
    
SplitFactTables()
```

### Assignment 2
Write a Python program that populates the database GroupIDHWMart with the
three fact tables you obtained from the previous assignments and with the necessary
linked dimensions described in the remaining .csv files. Note that in Time table the
values of the attributes day of week and quarter should be derived from the date.
Indicate each quarter with the values: Q1, Q2, Q3, Q4. While for each day of week
indicate each day with its proper name (“Monday”, “Tuesday”, etc.).

```python
import csv
import pyodbc
import datetime


server = 'tcp:apa.di.unipi.it'
username = 'group26' 
password = 'l3wwh' 
connectionString = 'DRIVER={ODBC Driver 17 for SQL Server};SERVER='+server+';UID='+username+';PWD='+ password


_db = pyodbc.connect(connectionString,autocommit=True)
_cursor = _db.cursor()

_cursor.fast_executemany = True



gpu_products = open('gpu.csv', 'r')
gpu_products_list = []
for row in csv.DictReader(gpu_products):
    gpu_products_list.append((int(row['gpu_code']), row['processor'], row['processor_manufacturer'], row['brand'], float(row['memory']), row['memory_type']))
gpu_products.close()
_cursor.executemany("INSERT INTO [Group26HWMart].[group26].[gpu_product] VALUES (?, ?, ?, ?, ?, ?)", gpu_products_list)
gpu_products_list = None
print('done1')

cpu_products = open('cpu.csv', 'r')
cpu_products_list = []
for row in csv.DictReader(cpu_products):
    cpu_products_list.append((int(row['cpu_code']), row['brand'], row['series'], row['name'], int(row['n_cores']), row['socket']))
cpu_products.close()
_cursor.executemany("INSERT INTO [Group26HWMart].[group26].[cpu_product] VALUES (?, ?, ?, ?, ?, ?)", cpu_products_list)
cpu_products_list = None
print('done2')

ram_product = open('ram.csv', 'r')
ram_product_list = []
for row in csv.DictReader(ram_product):
    ram_product_list.append((int(row['ram_code']), row['brand'], row['name'], float(row['memory']), row['memory_type'], int(row['clock'])))
ram_product.close()
_cursor.executemany("INSERT INTO [Group26HWMart].[group26].[ram_product] VALUES (?, ?, ?, ?, ?, ?)", ram_product_list)
ram_product_list = None
print('done3')

weekDays = ("Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday")
quarter = ['Q1', 'Q2', 'Q3', 'Q4']
time_file = open('time.csv')
time_list = []
for row in csv.DictReader(time_file):
    time_list.append((int(row['time_code']), int(row['year']), int(row['month']), int(row['day']), int(row['week']), quarter[(int(row['month'])-1) // 3], weekDays[datetime.date(int(row['year']), int(row['month']), int(row['day'])).weekday()]))
_cursor.executemany("INSERT INTO [Group26HWMart].[group26].[time] VALUES (?, ?, ?, ?, ?, ?, ?)", time_list)
time_file.close()
time_list = None
print('done4')

geography = open('geography.csv')
for row in csv.DictReader(geography):
    _cursor.execute("INSERT INTO [Group26HWMart].[group26].[geography] VALUES (?, ?, ?, ?, ?)", (int(row['geo_code']), row['continent'], row['country'], row['region'], row['currency']))
geography.close()
print('done5')

vendor = open('vendor.csv')
for row in csv.DictReader(vendor):
    _cursor.execute("INSERT INTO [Group26HWMart].[group26].[vendor] VALUES (?, ?)", (int(row['vendor_code']), row['name']))
vendor.close()
print('done6')
```
